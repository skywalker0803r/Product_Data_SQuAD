# MERGE DF AND DF_ERP

```python
df = df.rename(columns={'20':'LCNO'})
df3 = pd.merge(df,df_erp,on='LCNO',how='left')
```

```python
df3 = df3.dropna(subset=df_erp.columns)
```

```python
df3 = df3.drop_duplicates(subset=['LCNO'])
print(df3.shape)
df3.iloc[:,df3.columns.tolist().index('LCNO'):]
```

```python
df3[['45A']]
```

```python
import re

def preprocess(x):
    x = re.sub('[\u4e00-\u9fa5]', '', x) #去除中文
    x = re.sub('[’!"#$%&\'()*+,/:;<=>?@[\\]^_`{|}~，。,.]', '', x) #去除標點符號
    x = x.replace('\n', '').replace('\r', '').replace('\t', '') #去除換行符號
    x = str.strip(x) #移除左右空白
    return x

df3['45A'] = df3['45A'].apply(preprocess)
df3['SPEC'] = df3['SPEC'].apply(preprocess)
df3.loc[:,['45A','SPEC']]
```

```python
import numpy as np

寶典 = pd.read_excel('台塑企業_ 產品寶典20210303.xlsx')
產品集合 = set(寶典['品名'].values)
def Collection_method(df,產品集合):
    labels = {}
    for i in tqdm(df.index):
        products = []
        for p in 產品集合:
            if p in df.loc[i,'45A']:
                products.append(p)
        try:
            labels[i] = max(products,key=len) # 選最長的產品
        except:
            labels[i] = np.nan # 找不到就算了
    predict = pd.DataFrame(index=labels.keys(),columns=['Baodian_predict'])
    predict['Baodian_predict'] = labels.values()
    return predict
predict = Collection_method(df3,產品集合)
df3 = df3.join(predict)
df3
```

```python
ok_ = 0
no_ = 0
nan_ = 0
y_label = []

for i in tqdm(range(len(df3))):
    # 1.判斷SPEC是否匹配
    if df3['SPEC'].values[i] in df3['45A'].values[i]:
        y_label.append(df3['SPEC'].values[i])
        ok_ += 1
        continue
    
    # 2.判斷寶典是否匹配
    #try:
    #    if df3['Baodian_predict'].values[i] in df3['45A'].values[i]:
    #        y_label.append(df3['Baodian_predict'].values[i])
    #        ok_ += 1
    #        continue
    #except:
    #    nan_ += 1
    
    # 3.如果都不匹配 記起來
    no_ += 1
    y_label.append(np.nan)

    print(ok_,no_,nan_)
df3['Y_label'] = y_label
```

```python
def str2index(context,string):
    ys = context.find(string)
    ye = ys+len(string)
    return ys,ye
```

```python
ys,ye = str2index(df3['45A'].values[0],df3['Baodian_predict'].values[0])
df3['45A'].values[0][ys:ye]
```

```python
print(df3['Y_label'].isnull().sum())
df3 = df3.dropna(subset=['Y_label'],axis=0)
print(df3['Y_label'].isnull().sum())
```
