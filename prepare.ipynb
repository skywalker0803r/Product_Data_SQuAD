```python
import pandas as pd
import os
from tqdm import tqdm_notebook as tqdm
```

```python
lst = sorted(os.listdir())
lst
```

```python
cols = ['59','45A',
        '46A','47A',
        '31D','44C','47A','48','71D','78',
        '59','50','51A','45A','20']
```

```python
df = pd.DataFrame()
for l in tqdm(lst):
    try:
        d = pd.read_excel(l,sheet_name='mt700')[cols]
        df = df.append(d)
        print(df.shape)
    except:
        print(l)
        pass
```

```python
df = df.reset_index(drop=True)
df.head()
```

# ERP

```python
df_erp = pd.read_excel('../ERP資料/TB01.202101-06.xlsx',sheet_name='202101-06')
df_erp = df_erp[['LCNO','EXPNO','LCBK','CU1','SPEC']]
df_erp.head()
```

# MERGE DF AND DF_ERP

```python
df = df.rename(columns={'20':'LCNO'})
df3 = pd.merge(df,df_erp,on='LCNO',how='left')
```

```python
df3 = df3.dropna(subset=df_erp.columns)
```
