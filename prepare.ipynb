```python
import pandas as pd
import os
from tqdm import tqdm_notebook as tqdm
```

```python
lst = sorted(os.listdir('./電文資料'))
lst
```

```python
cols = ['59','45A',
        '46A','47A',
        '31D','44C',#'47A',
        '48','71D','78',
        #'59',
        '50','51A',#'45A',
        '20']
```

```python
df = pd.DataFrame()
for l in tqdm(lst):
    d = pd.read_excel('./電文資料/'+l,sheet_name='mt700')[cols]
    d['from'] = l
    df = df.append(d)
```

```python
df = df.reset_index(drop=True)
df.head()
```

# ERP

```python
df_erp = pd.read_excel('./ERP資料/TB01.202101-06.xlsx',sheet_name='202101-06')
df_erp = df_erp[['LCNO','EXPNO','LCBK','CU1','SPEC']]
df_erp['EXPNO'] = [ i[:2] for i in df_erp['EXPNO']]
df_erp.head()
```

# MERGE DF AND DF_ERP

```python
df = df.rename(columns={'20':'LCNO'})
df3 = pd.merge(df,df_erp,on='LCNO',how='left')
```

```python
df3 = df3.dropna(subset=df_erp.columns)
```

```python
df3 = df3.drop_duplicates()
df3
```

```python
df3.to_excel('combined_excel.xlsx')
```
